Data Cleaning in Pandas

```python
import pandas as pd
```

```python
# Load dataset
df = pd.read_excel(r"C:\Users\Leon\Downloads\Customer Call List.xlsx")
df
```

```python
# Remove duplicates and unnecessary columns
df = df.drop_duplicates()
df = df.drop(columns='Not_Useful_Column')
df
```

```python
# Clean Last_Name column
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

```python
# Clean and format Phone_Number column
df['Phone_Number'] = df['Phone_Number'].astype(str).str.replace(r'[-/|]', '', regex=True)
df['Phone_Number'] = df['Phone_Number'].apply(
    lambda x: x[:3] + '-' + x[3:6] + '-' + x[6:10] if len(x) == 10 else ''
)
df
```

```python
# Remove invalid phone number entries
df = df[df["Phone_Number"] != '']
df
```

```python
# Split Address column
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',', n=2, expand=True).fillna('')
df
```

```python
# Standardize Yes/No columns
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes', 'Y').str.replace('No', 'N')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes', 'Y').str.replace('No', 'N')
df
```

```python
# Remove rows where Do_Not_Contact is 'Y'
df = df[df["Do_Not_Contact"] != 'Y']
df
```

```python
# Handle missing values
df = df.replace('N/a', '').fillna('')
df
```

```python
# Reset index
df = df.reset_index(drop=True)
df
```
